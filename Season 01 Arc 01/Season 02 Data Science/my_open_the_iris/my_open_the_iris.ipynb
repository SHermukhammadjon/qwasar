```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix
from IPython.display import display
```

# 1.Loading the dataset.

```python
def load_dataset():
    return pd.read_csv("https://storage.googleapis.com/qwasar-public/track-ds/iris.csv")
```

```python
dataset = load_dataset()
# dataset
```

# 2.Summarizing the dataset.

```python
def summarize_dataset(dataset):
    #its shape
    print("Dataset dimension:")
    print(dataset.shape)

    #its ten first lines
    print("First 10 rows of dataset:")
    display(dataset.head(10))

    #its statistical summary
    print("Statistical summary:")
    display(dataset.describe())

    #Its distribution
    print("Class Distribution:")
    print(dataset.groupby('class').size())
```

```python
summarize_dataset(dataset)
```

# 4.Evaluating some algorithms.

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
def my_print_and_test_models(dataset):
    # split into X and Y
    X = dataset.drop("class" , axis=1)
    Y = dataset["class"]

    X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=0.20, random_state=1)
    
    # DecisionTree
    model = DecisionTreeClassifier()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"DecisionTree: {accuracy_score(X_predict, Y_validation)}")
    
    # GaussianNB
    model = GaussianNB()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"GaussianNB: {accuracy_score(X_predict, Y_validation)}")
    
    # KNeighbors
    model = KNeighborsClassifier()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"KNeighbors: {accuracy_score(X_predict, Y_validation)}")
    
    # LogisticRegression
    model = LogisticRegression()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"LogisticRegression: {accuracy_score(X_predict, Y_validation)}")
    
    # LinearDiscriminant
    model = LinearDiscriminantAnalysis()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"LinearDiscriminant: {accuracy_score(X_predict, Y_validation)}")
    
    # SVM
    model = SVC()
    model.fit(X_train, Y_train)
    
    X_predict = model.predict(X_validation)
    print(f"SVM: {accuracy_score(X_predict, Y_validation)}")
```

```python
my_print_and_test_models(dataset)
```
